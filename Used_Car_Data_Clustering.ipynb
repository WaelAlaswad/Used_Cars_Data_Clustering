# Data Splitting

```python
X = df.iloc[ : , 1:9].values
```

# Clustering By KMeans

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,10):
    kmeans = KMeans ( n_clusters=i ,init= "k-means++" ,random_state=42)
    kmeans.fit(X)

    wcss.append(kmeans.inertia_)

plt.plot(range(1,10) , wcss)
plt.xlabel(" Number Of Clusters")
plt.ylabel(" Inertia")
plt.title(" Elbow Graph")
plt.show()

```

```python
kmeans = KMeans ( n_clusters=3 ,init= "k-means++" ,random_state=42)
ypred = kmeans.fit_predict(X)
```

```python
plt.scatter(X[ypred == 0,0],X[ypred == 0,1], s= 20 , c= "red" , label= " Cluster 1")
plt.scatter(X[ypred == 1,0],X[ypred == 1,1], s= 20 , c= "blue" , label= " Cluster 2")
plt.scatter(X[ypred == 2,0],X[ypred == 2,1], s= 20 , c= "green" , label= " Cluster 3")



plt.scatter(kmeans.cluster_centers_[ : ,0], kmeans.cluster_centers_[: ,1], s=30 , c="yellow" , label = "Centers")
plt.show()
```

# Clustering By Hierarchy

```python
from scipy.cluster.hierarchy import linkage , dendrogram

dend = linkage( X , method="average",)

plt.figure(figsize=(30,10))
dendrogram(dend)
plt.show()
```

```python
dend = linkage( X , method="complete",)

plt.figure(figsize=(30,10))
dendrogram(dend)
plt.show()
```
